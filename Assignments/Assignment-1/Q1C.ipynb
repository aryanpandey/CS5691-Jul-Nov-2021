```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_plot_for_h(h):
    x = np.arange(0,10,0.01)
    y = []
    for k in x:
        if k<0:
            y.append(0)
        elif 0<=k<=1:
            y.append(1 - np.exp(-1*k/h))
        else:
            y.append(np.exp((1-k)/h) - np.exp(-k/h))
            
    plt.figure(figsize = (6,6))
    plt.plot(x,y)
    plt.title("h = " + str(h))
    plt.xlabel('x')
    plt.ylabel('E[p(x)]')
    plt.show()
    
get_plot_for_h(h = 1)
get_plot_for_h(h = 0.25)
get_plot_for_h(h = 0.0625)
```
